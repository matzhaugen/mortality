```python
import numpy as np

import matplotlib.pylab as plt
```

```python
period1 = 5
period2 = 2
def fn(t):
    return 1 + np.cos(2 * np.pi * t /period1) + np.cos(2 * np.pi * t / period2)
def fn2(t):
    return np.sin(2 * np.pi * t / period1) + np.cos(2 * np.pi * t / period2)
```

```python
from scipy.fft import fft, fftfreq
fig, axs = plt.subplots(5, 3, figsize=(9,8))

for k in np.arange(0, 5).astype(int):
    N = 300  # Number of sample points
    T = 1 / 10 * (k + 1)  # sample spacing
    t = np.linspace(0.0, N*T, N, endpoint=False)
    y = fn(t)
    axs[k, 0].plot(t, fn(t))
    axs[k, 0].set_xlim([0, 20])
    axs[k, 0].set_xlabel("time")
    
    p = plot_acf(fn(t), ax=axs[k,1], lags=20)
    axs[k, 1].set_xlabel("lag")
    
    yf = fft(y)
    xf = fftfreq(N, T)[:N//2]
    axs[k, 2].plot(xf, 2.0/N * np.abs(yf[0:N//2]))
    axs[k, 2].set_title("FFT")
    axs[k, 2].set_xlabel("frequency")
fig.tight_layout()
```
